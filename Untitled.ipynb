```python
import numpy as np
import cmath
from matplotlib import pyplot as plt
N=100
zmax=2

def mandelbrot(x, y):
    z=0
    c=complex(x, y)
    for i in range(N+1):
        z=z**2-c
        if np.abs(z)>=zmax:
            return cmath.phase(z)
        elif i==N:
            return cmath.phase(z)

x=np.linspace(-2.5, 2.5, 500)
y=np.linspace(-2.5, 2.5, 500)
z=[[0]*499 for i in range(499)]
for i in range(len(x)-1):
    for j in range(len(y)-1):
        z[i][j]=mandelbrot(x[i],y[j])
    
   
plt.pcolormesh(x, y, z, cmap='hsv')
plt.axis('scaled')
plt.show()
    
    
    
            
        
        
```
